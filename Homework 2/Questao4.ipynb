```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Define the transfer function parameters
k = 4800/7
num = k * np.array([1, 7])
den = [1, 25, 196, 480, 0]

# Create the transfer function
system = signal.TransferFunction(num, den)

# Define the time vector for the impulse response (inverse Laplace transform)
t = np.linspace(0, 2, 1000)  # 1000 points between 0 and 10 seconds

# Compute the impulse response (inverse Laplace transform)
t_imp, y_imp = signal.impulse(system, T=t)
t_stp, y_stp = signal.step(system, T=t)

# Plot the impulse response (time-domain representation)
plt.plot(t_imp, y_imp)
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.title('Impulse Response (Inverse Laplace Transform)')
plt.grid(True)
plt.show()

# Plot the impulse response (time-domain representation)
plt.plot(t_stp, y_stp)
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.title('Unit Step Response (Inverse Laplace Transform)')
plt.grid(True)
plt.show()
```
